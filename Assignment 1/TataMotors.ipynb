```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
# Load the data
tatamotors_df = pd.read_csv('TATAMOTORS.NS.csv')
nsei_df = pd.read_csv('^NSEI.csv')
```

```python

# Convert 'Date' to datetime and set as index
tatamotors_df['Date'] = pd.to_datetime(tatamotors_df['Date'])
nsei_df['Date'] = pd.to_datetime(nsei_df['Date'])
tatamotors_df.set_index('Date', inplace=True)
nsei_df.set_index('Date', inplace=True)
```

```python
# Calculate daily returns using Adjusted Close prices
tatamotors_df['Daily Return'] = tatamotors_df['Adj Close'].pct_change()
nsei_df['Daily Return'] = nsei_df['Adj Close'].pct_change()
```

```python
# Drop NaN values
tatamotors_df = tatamotors_df.dropna()
nsei_df = nsei_df.dropna()
```

```python
# computing alpha and beta

import statsmodels.api as sm

# Prepare the returns
X = nsei_df['Daily Return']
Y = tatamotors_df['Daily Return']
```

```python

# Add a constant to the market returns
X = sm.add_constant(X)

# Fit the regression model
model = sm.OLS(Y, X).fit()
alpha, beta = model.params
```

```python
alpha, beta
```

```python
# Annual sharpe ratio 
# Risk-free rate
risk_free_rate = 0.07015

# Calculate the daily risk-free rate
daily_risk_free_rate = (1 + risk_free_rate) ** (1 / 252) - 1

# Calculate the average daily return and standard deviation of daily returns
avg_daily_return_tata = tatamotors_df['Daily Return'].mean()
std_daily_return_tata = tatamotors_df['Daily Return'].std()

avg_daily_return_market = nsei_df['Daily Return'].mean()
std_daily_return_market = nsei_df['Daily Return'].std()

# Calculate the Sharpe ratios
sharpe_ratio_tata = (avg_daily_return_tata - daily_risk_free_rate) / std_daily_return_tata
sharpe_ratio_market = (avg_daily_return_market - daily_risk_free_rate) / std_daily_return_market

# Annualize the Sharpe ratios
annual_factor = np.sqrt(252)
annual_sharpe_ratio_tata = sharpe_ratio_tata * annual_factor
annual_sharpe_ratio_market = sharpe_ratio_market * annual_factor
```

```python
annual_sharpe_ratio_tata, annual_sharpe_ratio_market
```

```python
# Calculate Maximum Drawdown
def max_drawdown(returns):
    cumulative = (1 + returns).cumprod()
    peak = cumulative.cummax()
    drawdown = (cumulative - peak) / peak
    max_drawdown = drawdown.min()
    return max_drawdown

max_drawdown_tata = max_drawdown(tatamotors_df['Daily Return'])
max_drawdown_market = max_drawdown(nsei_df['Daily Return'])
```
