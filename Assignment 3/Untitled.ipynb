```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
data = pd.read_csv("fluidity.csv")
```

```python
data
```

```python
# Constants
INTEREST_RATE = 4.40 / 100

# Function to calculate the model price
def calculate_model_price(spot_price, expiry_date, current_date, interest_rate=INTEREST_RATE):
    T = (expiry_date - current_date).days / 365.0
    model_price = spot_price * np.exp(interest_rate * T)
    return model_price
```

```python
data.columns
```

```python
data.columns = data.columns.str.strip()
```

```python
# Correct column names based on inspection
expiry_col = 'Expiry'
date_col = 'Date'
future_price_col = 'Settle Price'
underlying_value_col = 'Underlying Value'
```

```python
# Convert date columns to datetime
data[expiry_col] = pd.to_datetime(data[expiry_col])
data[date_col] = pd.to_datetime(data[date_col])

# Calculate Settle Price (assumed to be the same as Model Price)
data['Settle Price'] = data[future_price_col]

# Calculate Model Price
data['Model Price'] = data.apply(lambda row: calculate_model_price(row[underlying_value_col], row[expiry_col], row[date_col]), axis=1)

# Spot Price is the same as the Underlying value
data['Spot Price'] = data[underlying_value_col]
```

```python

# Create output DataFrame
output_data = data[[date_col, 'Settle Price', 'Model Price', 'Spot Price']]
```

```python
output_file = 'output_metrics.csv'
output_data.to_csv(output_file, index=False)
```

```python

# Plot the graph
plt.figure(figsize=(10, 6))
plt.plot(data[date_col], data['Settle Price'], label='Settle Price', marker='o')
plt.plot(data[date_col], data['Model Price'], label='Model Price', marker='x')
plt.plot(data[date_col], data['Spot Price'], label='Spot Price', marker='s')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Settle Price, Model Price, and Spot Price over Time')
plt.legend()
plt.grid(True)
plt.savefig('price_graph.png')
plt.show()
```

```python
# Calculate the mean of the interest rates obtained for all data points
import matplotlib.pyplot as plt
from scipy import stats

def calculate_implied_interest_rate(row):
    time_diff = (row[expiry_col] - row[date_col]).days / 365.0
    if time_diff > 0:
        return np.log(row['Settle Price'] / row['Spot Price']) / time_diff
    else:
        return np.nan

data['Implied Interest Rate'] = data.apply(calculate_implied_interest_rate, axis=1)

# Print out the implied interest rates for debugging
print(data[['Date', 'Expiry', 'Settle Price', 'Spot Price', 'Implied Interest Rate']])

# Filter out rows with NaN implied interest rates
filtered_data = data.dropna(subset=['Implied Interest Rate'])


z_scores = stats.zscore(filtered_data['Implied Interest Rate'])
filtered_data = filtered_data[(z_scores < 3) & (z_scores > -3)]

mean_implied_interest_rate = filtered_data['Implied Interest Rate'].mean()

# Calculate the percentage error
percentage_error = abs((mean_implied_interest_rate - INTEREST_RATE) / INTEREST_RATE) * 100
```

```python

# Output the results
print(f'Mean Implied Interest Rate: {mean_implied_interest_rate:.4f}')
print(f'Given Interest Rate: {INTEREST_RATE:.4f}')
print(f'Percentage Error: {percentage_error:.2f}%')

# Optionally save the mean implied interest rate and percentage error to a file
with open('interest_rate_comparison.txt', 'w') as f:  # Specify your desired path here or comment out if not needed
    f.write(f'Mean Implied Interest Rate: {mean_implied_interest_rate:.4f}\n')
    f.write(f'Given Interest Rate: {INTEREST_RATE:.4f}\n')
    f.write(f'Percentage Error: {percentage_error:.2f}%\n')
```
